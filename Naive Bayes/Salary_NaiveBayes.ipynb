```python
data_train = pd.read_csv('SalaryData_Train.csv')
data_train
```

```python
data_train.info()
```

```python
data_test = pd.read_csv('SalaryData_Test.csv')
data_test
```

```python
data_test.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
select_columns=["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]
for i in select_columns:
    data_train[i] = le.fit_transform(data_train[i])
    data_test[i] = le.fit_transform(data_test[i])

colnames = data_train.columns
```

```python
data_train
```

```python
data_test
```

```python
x_train = data_train.drop(['Salary'],axis=1)
y_train = data_train['Salary']
```

```python
x_test = data_test.drop(['Salary'],axis=1)
y_test = data_test['Salary']
```

```python
# Normalization Function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
train_norm = norm_func(x_train)
train_norm
```

```python
test_norm = norm_func(x_test)
test_norm
```

```python
from sklearn.naive_bayes import CategoricalNB

model = CategoricalNB()

model.fit(train_norm,y_train)
```

```python
pred_train = model.predict(train_norm)
```

```python
np.mean(pred_train==y_train)
```

```python
pred_test = model.predict(test_norm)
pred_test
```

```python
pd.crosstab(y_test,pred_test)
```

```python
np.mean(pred_test==y_test)
```
